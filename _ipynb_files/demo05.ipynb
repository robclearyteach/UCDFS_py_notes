### Tuple immutability (un-changeable ~ after creation)

```python
bus = (14, "dundrum")   #immutable
bus[1] = "cabinteely"
```

```python
aa   = (14, "dundrum")
bb   = (14, "dundrum")

print( id(aa), id(bb) )
print( aa is bb)
print( aa == bb)
```

```python
aa   = (14, "dundrum")
bb   = aa                      ##bb is an 'alias' for aa

print( id(aa), id(bb) )
print( aa is bb)
print( aa == bb)
```

```python
a = (14, "dundrum")     #immutable
b = a                   #change to a new tuple object
```

### Tuple un-packing

```python
#re-cap tuple un-packing
bus = (14, "dundrum")
number, destination = bus       #un-pack
(number, destination) = bus

print(number)
print(destination)
```

```python
other_fruits = ["pear", "grape", "pineapple", "watermelon"]
fruits       = ["apple", "orange", "tomato", "banana"]

orig_fruits = fruits
print( id(fruits), fruits ) 
print( id(orig_fruits), orig_fruits ) 

fruits = fruits + other_fruits

print( id(fruits), fruits ) 
print( id(orig_fruits), orig_fruits ) 
#["apple", "orange", "tomato", "banana","pear", "grape", "pineapple", "watermelon"]
```

### repeating question loop

```python
# without a loop
inputs = []

response = input( "Ask: enter a num" )
inputs.append(response)

response = input( "Ask: enter a num" )
inputs.append(response)

        
#how to get to here in the execution
print(inputs)
```

```python
inputs = []

response = input( "Ask: enter a num" )
while response:
    inputs.append(response)
    response = input( "Ask: enter a num" )

print(inputs)
```
