### Tuple immutability (un-changeable ~ after creation)

```python
bus = (14, "dundrum")   #immutable
bus[1] = "cabinteely"
```

```python
bus_route_14 = (14, "dundrum")   #immutable
print( id(bus_route_14) )
# bus_route_14[1] = "tallaght"   # can't assign an element

bus_route_14 = (14, "tallaght")     #change to a new tuple object
print( id(bus_route_14) )
```

```python
# a way around tuple immutability
bus_route_14   = (14, "dundrum")
list_bus_route =  list(bus_route_14) 

print(list_bus_route)
list_bus_route[1] = "tallaght"
print(list_bus_route)

bus_route_14   = tuple(list_bus_route)

print( bus_route_14)
```

```python
#Be aware: tuples are immutable
# if an element of a tuple is mutable
# you can't replace it but you can change the 
# contents of an element itself
# (it the element is mutable like a list)
```

```python
sports_data = (37.0, [140, 137])

sports_data[1][0] = 143
print(sports_data[1][1])
```

```python
sports_data = (37.0, [140, 137])
sports_data[1] = [143, 137]
```

### Tuple un-packing

```python
#re-cap tuple un-packing
bus = (14, "dundrum")
number, destination = bus       #un-pack
(number, destination) = bus

print(number)
print(destination)
```
