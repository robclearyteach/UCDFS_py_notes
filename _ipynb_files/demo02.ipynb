```python
### Previously...
import time

# Get the current time in seconds since the epoch
now = time.time()
now = int(now)  #don't need factions of seconds
seconds_since_epoch = now
print( f"{seconds_since_epoch=}")

# Calculate total seconds in a day
seconds_in_a_day = (60 * 60) * 24 
print(f"{seconds_in_a_day=}")

# Find the number of seconds that have passed today
seconds_today =  seconds_since_epoch % seconds_in_a_day
print(f"{seconds_today=}")
seconds_in_an_hour = 60 * 60
# Calculate the current hour, minute, and second
hours   = seconds_today//seconds_in_an_hour
minutes = (seconds_today% seconds_in_an_hour )// 60
seconds = seconds_today%60


# Print the time in HH:MM:SS format
print(f"{hours:02}:{minutes:02}:{seconds:02}")
```
