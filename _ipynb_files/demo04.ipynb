## Python conditional 'ternary' expression

```python
#Observe the workings of this
# conditional expression
# TASK: Say/write-out/express what it does?
what_to_do = None                           #None => no value yet

have_energy = True

if have_energy:
    what_to_do = "go on a trip"
else:
    what_to_do =  "watch TV"


print(f"Let's {what_to_do}")
```

```python
#Observe this re-working of the above
# What is the difference: 
#   it's a function and this time it...

def what_to_do( have_energy ):
    if have_energy:
        what_to_do = "go on a trip"
        return what_to_do
    else:
        what_to_do =  "watch TV"
        return what_to_do

what_to_do_returned = what_to_do( True )            #True => have_energy
print(f"Let's {what_to_do_returned}")
```

## Python 'match' statement (or 'case' statement)

```python
status = 408

match status:
    case 400:
        print("Bad request")
    case 404:
        print("Not found")
    case 418:
        print("I'm a teapot")
    case _:
        print("Something's wrong with the internet")
```

```python
status = 401    #Try: 402  #403

match status:
    case 400:
        print("Bad request")
    case 401 | 403 | 404:
        print("Not allowed")
    case 404:
        print("Not found")
    case 418:
        print("I'm a teapot")
    case _:
        print("Something's wrong with the internet")
```

```python
check_ = [3,4]
a=3
b=4

match check_:
    case 0:
        print("zero")
    case 1 | 2:
        print("one or two")
    case [a, b]:
        print(f"a list with two items: {a}, {b}")
    case _:
        print("something else")
```

```python
check_ = 'test'


match check_:
    case 0:
        print("zero")
    case 1 | 2:
        print("one or two")
    case [a, b]:
        print(f"a list with two items: {a}, {b}")
    case 'test':                                                #added String 'test'
        print(f"a string 'test' ")
    case _:
        print("something else")
```

```python
#Python match 'case'
#   direct expressions not allowed
def get_val():
    return 3

check_ = 10

match check_:
    case 5 + 5:          # expression gives SyntaxError
        print("matched 10")
    case 2 * 3:
        print("matched 6")
    case _:
        print("no match")
```

```python

def get_val():
    return 3

check_ = 10

match check_:
    case x if x == 5 + 5:
        print("matched 10")
    case x if x == 2 * 3:
        print("matched 6")
    case _:
        print("no match")
```

```python
check_ = 'hungry'       #67 #bread      #try other values

match check_:
    case "hello":                       # literal string
        print("greeting")
    case s if s.startswith("h"):        # guard with binding
        print("starts with h:", s)
    case str():                         # class pattern
        print("some other string")
    case _:
        print("not a string")
```

```python
point = (0, 2)
y = 2

match point:
    case (0, 0):
        print("Origin")
    case (0, y):
        print(f"Y={y}")
    case (x, 0):
        print(f"X={x}")
    case (x, y):
        print(f"X={x}, Y={y}")
    case _:
        raise ValueError("Not a point")
```

### Body Mass Index (BMI) basic calculator

```python
#Suppose known constant values that shouldn't change...
KG_PER_POUND = 0.45359237   # Py `Constant`
MTR_PER_INCH = 0.0254       # Py `Constant`

#BUT...
KG_PER_POUND = 7            #can still change it
print( KG_PER_POUND)
```

```python
#lec03_eg04
"""
Body Mass Index (BMI) is a measure of health on weight. 
It can be calculated by taking your weight in kilograms and 
dividing by the square of your height in meters. 
The interpretation of BMI for people 16 years or older is as follows:
+-------------+-----------------------+
| BMI         | Category              |
+-------------+-----------------------+
| < 18.5      | Underweight           |
| 18.5 - 24.9 | Normal weight         |
| 25 - 29.9   | Overweight            |
| >= 30       | Weight Issue (Class 1)|
+-------------+-----------------------+
"""
def KG_PER_POUND():
    return 0.45359237   # Constant

def MTR_PER_INCH():
    return 0.0254       # Constant 

# weight_lbs = float( input("Enter weight in pounds:"))
# height_in  = float( input("Enter your height in inches:") )
weight_lbs = 170
height_in  = 55
 
weight_kgs = weight_lbs * KG_PER_POUND()
height_mtr = height_in  * MTR_PER_INCH()
heigth_mtr_squared = height_mtr**2             #math -> pow( height_mtr, 2)

bmi = weight_kgs/heigth_mtr_squared
msg = "unknown"

if   bmi <= 18.5:
    msg = "Underweight"
elif bmi < 25:
    msg = "Normal weight"
elif bmi < 30:
    msg = "Overweight"
else:
    msg = "Weight Issue (Class 1)"

print(f"your BMI is: {bmi}: the BMI category states: {msg}")
```

### Subtraction Tutor

```python
#lec03_eg03
# Subtraction tutor: suppose want it to ask 5 sums
import random

num1 = random.randint(0, 10)
num2 = random.randint(0, 10)

if num1 > num2:
    bigger, smaller = num1, num2
else:
    bigger, smaller = num2, num1

print( f"what is {bigger} minus {smaller}?" )
user_input = int( input("what is the answer?") )
correct_answer = bigger-smaller
if( user_input == correct_answer):
    print("Correct! well done")
else: 
    print(f"{bigger} minus {smaller} is {correct_answer}: try again!")
```

```python
#Suppose I want a quiz with 5 sums:

for _ in range(5):
    num1 = random.randint(0, 10)
    num2 = random.randint(0, 10)

    if num1 > num2:
        bigger, smaller = num1, num2
    else:
        bigger, smaller = num2, num1

    print( f"what is {bigger} minus {smaller}?" )
    user_input = int( input("what is the answer?") )
    correct_answer = bigger-smaller
    if( user_input == correct_answer):
        print("Correct! well done")
    else: 
        print(f"{bigger} minus {smaller} is {correct_answer}: try again!")
```
