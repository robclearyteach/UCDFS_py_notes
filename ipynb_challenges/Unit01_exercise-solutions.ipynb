```python
# 3. Display a pattern.
print("""
FFFFFFF   U     U   NN     NN
FF        U     U   NNN    NN
FFFFFFF   U     U   NN N   NN
FF         U   U    NN  N  NN
FF          UUU     NN    NNN
""")
```

```python
# 4. Print a table.
# We could just use a regular multiline string:
print("""
a   a^2   a^3
1   1     1
2   4     8
3   9     27
4   16    64
""")
```

```python
# Another option is to use the tab character \t , to not have to worry about aligning the columns:
print("""
a\ta^2\ta^3
1\t1\t1
2\t4\t8
3\t9\t27
4\t16\t64
""")
```

```python
# We could also actually use an f-string, for the calculations:
print(f"a\ta^2\ta^3")
print(f"1\t{1**2}\t{1**3}")
print(f"2\t{2**2}\t{2**3}")
print(f"3\t{3**2}\t{3**3}")
print(f"4\t{4**2}\t{4**3}")

# And in the coming lessons you'll see how to do this with loops
```
